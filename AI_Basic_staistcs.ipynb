```python
import matplotlib.pyplot as plt

# Data points and their positions
positions = [2, 3, 7]
weights = [1, 1, 1]  # Equal weights

# Calculate the mean (balance point)
mean_position = sum(p * w for p, w in zip(positions, weights)) / sum(weights)

# Plot setup
fig, ax = plt.subplots()
ax.set_xlim(0, 9)
ax.set_ylim(-1, 2)
ax.axhline(0, color='black', linewidth=2)  # Number line

# Plot points
for pos in positions:
    ax.plot(pos, 0.1, 'ko', markersize=10)
    ax.annotate(f'{pos}', (pos, -0.3), ha='center', fontsize=14)

# Draw triangle at the mean (balance point)
ax.plot(mean_position, -.1, marker='^', color='red', markersize=15)
ax.annotate(f'{mean_position:.1f}', (mean_position, -.5), color='red', ha='center', fontsize=14)

# Hide axes
ax.axis('off')
plt.title("Mean as Balance Point", fontsize=16)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Initial data points
positions = [2, 3, 7, 8]

plt.ion()  # Turn on interactive mode

while True:
    # Equal weights
    weights = [1] * len(positions)

    # Calculate the mean (balance point)
    mean_position = sum(p * w for p, w in zip(positions, weights)) / sum(weights)

    # Plot setup
    fig, ax = plt.subplots()
    ax.set_xlim(0, max(positions) + 1)
    ax.set_ylim(-1, 2)
    ax.axhline(0, color='black', linewidth=2)  # Number line

    # Plot points
    for pos in positions:
        ax.plot(pos, 0.1, 'ko', markersize=10)
        ax.annotate(f'{pos}', (pos, -0.3), ha='center', fontsize=14)

    # Draw triangle at the mean (balance point)
    ax.plot(mean_position, -0.1, marker='^', color='red', markersize=15)
    ax.annotate(f'{mean_position:.1f}', (mean_position, -0.5), color='red',ha='center', fontsize=14)

    # Hide axes
    ax.axis('off')
    plt.title("Mean as Balance Point", fontsize=16)
    plt.show()
    plt.pause(0.1)  # Pause to render figure

    # Ask for user input AFTER figure is shown
    user_input = input("Enter a new position to add (or type 'stop' to finish): ").strip()

    plt.close()  # Close previous figure

    if user_input.lower() == 'stop':
        print("Exiting.")
        break

    try:
        new_position = float(user_input)
        positions.append(new_position)
    except ValueError:
        print("Invalid input! Please enter a number or 'stop'.")
```

```python
import matplotlib.pyplot as plt
import time

# Initial positions
initial_positions = [2, 3, 7, 8]

# Ask the user for additional positions (comma-separated)
user_input = input("Enter additional positions separated by commas (e.g., 5, 9, 6): ")
try:
    additional_positions = [float(x.strip()) for x in user_input.split(',') if x.strip()]
except ValueError:
    print("Invalid input! Please enter numbers separated by commas.")
    exit()

# Combine initial and new positions
all_positions = initial_positions + additional_positions
positions_so_far = []

plt.ion()  # Enable interactive mode

for i in range(len(all_positions)):
    positions_so_far.append(all_positions[i])
    weights = [1] * len(positions_so_far)
    mean_position = sum(p * w for p, w in zip(positions_so_far, weights)) / sum(weights)

    # Plot
    fig, ax = plt.subplots()
    ax.set_xlim(0, max(all_positions) + 1)
    ax.set_ylim(-1, 2)
    ax.axhline(0, color='black', linewidth=2)

    # Plot points so far
    for pos in positions_so_far:
        ax.plot(pos, 0.1, 'ko', markersize=10)
        ax.annotate(f'{pos}', (pos, -0.3), ha='center', fontsize=14)

    # Plot mean
    ax.plot(mean_position, -0.1, marker='^', color='red', markersize=15)
    ax.annotate(f'Mean: {mean_position:.2f}', (mean_position, -0.5), ha='center', fontsize=14)

    ax.axis('off')
    plt.title(f"Step {i+1}: Mean as Balance Point", fontsize=16)
    plt.show()
    plt.pause(1.0)  # Pause to view animation frame
    plt.close()
```

```python
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.figure import Figure

# Prepare the data
initial_positions = [2, 3, 7, 8]
additional_positions = [10, 16, 20, 4]  # Example values
all_positions = initial_positions + additional_positions

# Create a figure and axis
fig, ax = plt.subplots()

# Function to update the frame
def update(frame):
    ax.clear()
    positions_so_far = all_positions[:frame + 1]
    weights = [1] * len(positions_so_far)
    mean_position = sum(p * w for p, w in zip(positions_so_far, weights)) / sum(weights)

    # Set limits and plot number line
    ax.set_xlim(0, max(all_positions) + 1)
    ax.set_ylim(-1, 2)
    ax.axhline(0, color='black', linewidth=2)

    # Plot points
    for pos in positions_so_far:
        ax.plot(pos, 0.1, 'ko', markersize=10)
        ax.annotate(f'{pos}', (pos, -0.3), ha='center', fontsize=14)

    # Plot mean
    ax.plot(mean_position, -0.1, marker='^', color='red', markersize=15)
    ax.annotate(f'Mean: {mean_position:.2f}', (mean_position, -0.5), ha='center', fontsize=14)

    ax.axis('off')
    ax.set_title(f"Step {frame + 1}: Mean as Balance Point", fontsize=16)

# Create the animation
ani = animation.FuncAnimation(fig, update, frames=len(all_positions), repeat=False)

# Save as GIF
ani.save('stat.gif', fps=1)

```
